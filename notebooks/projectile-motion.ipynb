```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python

plt.figure(dpi=200)

colors = ['#0080ff', '#2090ff', '#40a0ff', '#60b0ff', '#80c0ff', '#aad0ff']

g = 9.81
a = np.array([0, -g])
angle = math.pi / 4
force = 100
vx = force * math.cos(angle)
vy = force * math.sin(angle)
v = np.array([vx, vy])
p = np.array([0, 0])
T = np.linspace(0, 2 * force * math.sin(angle) / g, 20)
x, y = zip(*[0.5 * a * (t * t) + v * t for t in T])

plt.plot(x, y, c=colors[0], lw=2)
plt.scatter(x, y, c=colors[0], s=10)

plt.axhline(0, c='gray', ls='dashed', alpha=0.5)
plt.gca().set_aspect('equal')
# plt.legend(['Projectile'])
plt.xlabel('Horizontal distance (m)')
plt.ylabel('Vertical distance (m)')
```
